```python
def main():
    """
    Main method to run.
    """
    perceptron_training()
    
    
def perceptron_training():
    activation_function = lambda x: 0 if x < 0 else 1
    data = [[0, 0], [0, 1], [1, 0], [1, 1]]

    operators = [("AND", [0, 0, 0, 1]), ("OR", [0, 1, 1, 1]), ("XOR", [0, 1, 1, 0])]

    for operator, labels in operators:
        perceptron = Perceptron(len(data[0]), activation_function)
        perceptron.train(data, labels, operator)

if __name__ == "__main__":
    main()
```
